```python
file_path = '/gdrive/MyDrive/NLP/week02/2-2. POS Tagged Corpus for Zipf Law.txt'

dic_rank={}
dic_frequency ={}
with open(file_path, 'r', encoding="euc-kr") as inFile:
  lines = inFile.readlines()

for line in lines:
  line=line.strip()
  _, words = line.split('\t')
  mophology = words.split('+')
  for m in mophology:
    if m not in dic_frequency:
      dic_frequency[m] = 1
    else:
      dic_frequency[m] += 1

# 딕셔너리를 값에 따라 오름차순으로 정렬
sorted_dict = dict(sorted(dic_frequency.items(), key=lambda item: item[1], reverse=True))

# 값의 크기에 따라 순위 부여
rank = 1
for key, _ in sorted_dict.items():
  dic_rank[key]=rank
  rank+=1

# 순위를 기준으로 키를 내림차순 정렬하여 최종 결과 생성
final_dic_rank = dict(sorted(dic_rank.items(), key=lambda item: item[1]))

count=0
print(f"형태소/품사, 빈도수, 순위*빈도수\n")
for word in final_dic_rank:
  if count==100:
    break
  print(f"{word}, {dic_frequency[word]}, {final_dic_rank[word]*dic_frequency[word]}")
  count+=1
```
